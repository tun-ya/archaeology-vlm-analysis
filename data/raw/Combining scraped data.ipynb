```python
import pandas as pd
```

```python
df = pd.read_csv('./notebooks/merged_file.csv')
```

```python
json_df = pd.read_json('./notebooks/combined_all_artifacts.json')
```

```python
# Check the number of rows and columns
num_rows, num_columns =json_df.shape
```

```python
num_rows
```

```python
# Step 1: Merge the two DataFrames on image_id (df) and object_number (json_df)
# Use a left merge to retain all rows from df
merged_df = pd.merge(df, json_df, left_on='image_id', right_on='object_number', how='left')

# Step 2: Update the artifact_title in df with the values from the merged json_df
# Fill the missing values in the artifact_title column with the new values from json_df
merged_df['artifact_title'] = merged_df['artifact_title_y'].combine_first(merged_df['artifact_title_x'])

# Drop unnecessary columns after merge
merged_df = merged_df.drop(columns=['artifact_title_x', 'artifact_title_y', 'object_number', 'object_number_x', 'object_number_y'])

merged_df = merged_df.drop_duplicates(subset=['image_id'])

# Step 3: Save the updated DataFrame to a CSV file
merged_df.to_csv('updated_artifacts.csv', index=False)

# Display the updated DataFrame
print(merged_df)
```
